```python
def full_pyramid(n):
    for i in range(1, n + 1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")

        # Print asterisks for the current row
        for k in range(1, 2*i):
            print("*", end="")
        print()

full_pyramid(5)
```

Example 3: Pyramid Patterns in Python with Number

```python
def print_number_pyramid(rows):
    for i in range(1, rows + 1):
        # Print spaces
        for j in range(rows - i):
            print(" ", end="")
        # Print numbers
        for j in range(2 * i - 1):
            print(j + 1, end="")
        # Move to the next line after each row
        print()

# Example usage
num_rows = 5
print_number_pyramid(num_rows)
```

Example 4: Inverted Full Pyramid Patterns in Python





```python
# Function to print inverted full pyramid pattern
def inverted_full_pyramid(n):
    # Outer loop for the number of rows
    for i in range(n, 0, -1):
        # Inner loop for leading spaces in each row
        for j in range(n - i):
            print(" ", end="")
        # Inner loop for printing asterisks in each row
        for k in range(2*i - 1):
            print("*", end="")
        # Move to the next line after each row
        print("")

# Set the value of n (number of rows)
n = 5
 #Call the function to print the inverted full pyramid
inverted_full_pyramid(n)
```

Example 5: Hollow Pyramid Patterns in Python

```python
def hollow_pyramid(n):
    for i in range(1, n + 1):
        for j in range(1, 2 * n):
            if j == n - i + 1 or j == n + i - 1 or i == n:
                print("*", end="")
            else:
                print(" ", end="")
        print()

# Set the number of rows for the pyramid
rows = 5

# Call the function with the specified number of rows
hollow_pyramid(rows)
```

BankAccount class banao jisme __balance private ho, deposit() aur withdraw() methods ho.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdraw amount must be positive")
        elif amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")

    def get_balance(self):
        return self.__balance
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)

print("Current Balance:", account.get_balance())
```

@property ka use karke ek private variable ko readâ€“write banao with validation.

```python
class Person:
    def __init__(self, age):
        self.__age = age   # private variable

    @property
    def age(self):
        # getter
        return self.__age

    @age.setter
    def age(self, value):
        # setter with validation
        if value < 0:
            print("Age cannot be negative")
        else:
            self.__age = value
p = Person(25)

print(p.age)     # âœ… read (getter)

p.age = 30       # âœ… write (setter)
print(p.age)

p.age = -5       # âŒ invalid
```

LibraryBook class banao jisme __is_issued private ho aur issue_book() & return_book() methods ho.

```python
class LibraryBook:
    def __init__(self):
        self.__is_issued = False   # private variable

    def issue_book(self):
        if self.__is_issued:
            print("Book is already issued")
        else:
            self.__is_issued = True
            print("Book issued successfully")

    def return_book(self):
        if not self.__is_issued:
            print("Book is not issued yet")
        else:
            self.__is_issued = False
            print("Book returned successfully")
book = LibraryBook()

book.issue_book()    # issue
book.issue_book()    # already issued

book.return_book()   # return
book.return_book()   # not issued
```

```python
class ATM:
    def __init__(self, cash_available):
        self.__cash_available = cash_available   # private
        self.__withdrawal_limit = 20000          # per transaction

    def withdraw(self, amount):
        if amount <= 0:
            print("Enter a valid amount")
        elif amount > self.__withdrawal_limit:
            print("Withdrawal limit exceeded (Max â‚¹20,000)")
        elif amount > self.__cash_available:
            print("ATM has insufficient cash")
        else:
            self.__cash_available -= amount
            print(f"Please collect your cash: â‚¹{amount}")

    def get_cash_available(self):
        return self.__cash_available
atm = ATM(50000)

atm.withdraw(10000)   # success
atm.withdraw(30000)   # limit exceeded
atm.withdraw(50000)   # insufficient cash

print("ATM Cash Left:", atm.get_cash_available())
```

MovieTicket class banao jisme __seat_number private ho aur double booking na ho.

```python
class MovieTicket:
    def __init__(self, seat_number):
        self.__seat_number = seat_number   # private
        self.__is_booked = False

    def book_ticket(self):
        if self.__is_booked:
            print(f"Seat {self.__seat_number} is already booked")
        else:
            self.__is_booked = True
            print(f"Seat {self.__seat_number} booked successfully")

    def cancel_ticket(self):
        if not self.__is_booked:
            print(f"Seat {self.__seat_number} is not booked yet")
        else:
            self.__is_booked = False
            print(f"Booking for seat {self.__seat_number} cancelled")
ticket = MovieTicket(12)

ticket.book_ticket()     # book
ticket.book_ticket()     # double booking attempt

ticket.cancel_ticket()   # cancel
ticket.book_ticket()     # book again
```

FuelTank class banao jisme __fuel_level private ho aur overfill na ho.

```python
class FuelTank:
    def __init__(self, capacity):
        self.__capacity = capacity
        self.__fuel_level = 0   # private

    def fill_fuel(self, amount):
        if amount <= 0:
            print("Enter a valid fuel amount")
        elif self.__fuel_level + amount > self.__capacity:
            print("Cannot overfill the tank")
        else:
            self.__fuel_level += amount
            print(f"Fuel filled: {amount} liters")

    def use_fuel(self, amount):
        if amount <= 0:
            print("Enter a valid fuel amount")
        elif amount > self.__fuel_level:
            print("Not enough fuel")
        else:
            self.__fuel_level -= amount
            print(f"Fuel used: {amount} liters")

    def get_fuel_level(self):
        return self.__fuel_level
tank = FuelTank(50)

tank.fill_fuel(30)    # ok
tank.fill_fuel(25)    # overfill attempt

tank.use_fuel(10)
print("Fuel left:", tank.get_fuel_level())
```

OnlineOrder class banao jisme private order status ho aur direct change na ho.

```python
class OnlineOrder:
    def __init__(self):
        self.__order_status = "Placed"   # private

    def ship_order(self):
        if self.__order_status == "Placed":
            self.__order_status = "Shipped"
            print("Order shipped successfully")
        else:
            print("Order cannot be shipped")

    def deliver_order(self):
        if self.__order_status == "Shipped":
            self.__order_status = "Delivered"
            print("Order delivered successfully")
        else:
            print("Order cannot be delivered")

    def get_order_status(self):
        return self.__order_status
order = OnlineOrder()

print(order.get_order_status())  # Placed

order.ship_order()
print(order.get_order_status())  # Shipped

order.deliver_order()
print(order.get_order_status())  # Delivered
```

Ek abstract class FileHandler banao jisme read() aur write() methods ho.

```python
from abc import ABC, abstractmethod

class FileHandler(ABC):

    @abstractmethod
    def read(self):
        pass

    @abstractmethod
    def write(self, data):
        pass
class TextFile(FileHandler):
    def read(self):
        return "Reading text file"

    def write(self, data):
        return f"Writing '{data}' to text file"
class BinaryFile(FileHandler):
    def read(self):
        return "Reading binary file"

    def write(self, data):
        return f"Writing binary data '{data}'"
t = TextFile()
b = BinaryFile()

print(t.read())
print(t.write("Hello"))

print(b.read())
print(b.write("101010"))
```
