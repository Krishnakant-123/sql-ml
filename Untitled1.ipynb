### Switching to SQLite

Since direct connection to a local MySQL server from Google Colab is not straightforward, we will use **SQLite**. SQLite is a file-based database that doesn't require a separate server process, making it ideal for Colab environments.

We will create a database file named `college.db`.

```python
import sqlite3

# Connect to an SQLite database (it will create the file if it doesn't exist)
conn = sqlite3.connect('college.db')
cursor = conn.cursor()

print("Connected to SQLite database 'college.db' successfully.")
```

Now, let's recreate the `students` table and insert some data using SQLite.

```python
# Drop the table if it already exists (useful for re-running the cell)
cursor.execute('DROP TABLE IF EXISTS students;')

# Create the students table
cursor.execute('''
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name VARCHAR(50),
    age INTEGER
);
''')

# Insert values
cursor.execute("INSERT INTO students (id, name, age) VALUES (1, 'rahul', 20)")
cursor.execute("INSERT INTO students (id, name) VALUES (2, 'amit')")

# Commit the changes
conn.commit()

print("Table 'students' created and data inserted successfully.")
```

You can now query the `students` table.

```python
# Select all data from students
cursor.execute('SELECT * FROM students;')
rows = cursor.fetchall()
print("Students table data:")
for row in rows:
    print(row)

# Select specific columns
cursor.execute('SELECT name, age FROM students;')
rows = cursor.fetchall()
print("\nStudents names and ages:")
for row in rows:
    print(row)

# Select with WHERE clause
cursor.execute('SELECT * FROM students WHERE age > 18;')
rows = cursor.fetchall()
print("\nStudents older than 18:")
for row in rows:
    print(row)
```

Don't forget to close the connection when you are done.

```python
# Close the connection
conn.close()
print("Connection to SQLite database closed.")
```
